
## Exploratory Data Analysis 



### Gender

```python
ax=sns.countplot(x='Gender',data=df)
```

```python
ax = sns.countplot( x = 'Gender' , data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(["Gender"])["Amount"].sum()
```

```python
total = df.groupby( ['Gender'] , as_index = False ) ['Amount'].sum().sort_values( by = 'Amount')

sns.barplot( x = 'Gender' , y = 'Amount' , data = total)
```




##### From above graphs we can observe that most of the buyers are  female , and even the purschasing power of female is greater than men







### Age Group

```python
ax = sns.countplot(x="Age Group" , data=df , hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Age Group'])['Amount'].sum()
```

```python
df.groupby ( ['Age Group'] , as_index=False ) ['Amount'].sum().sort_values(by = 'Amount')
```

```python
Group =  df.groupby(["Age Group"] , as_index=False)["Amount"].sum().sort_values(by = "Amount")

sns.barplot(x='Age Group' , y="Amount" , data=Group)
```

###### From the above graph we can observe that most of the buyers are from age group between 26-35 yrs female

## State

```python
df.groupby(["State"])["Orders"].sum().sort_values()
```

```python
sales_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values( by = "Orders" , ascending=False ).head(5)

sns.barplot(x='State' , y='Orders' , data = sales_state)
```

##### From the above graphs we can observe that most of the orders are from uttar pradesh and maharashtra

## Marital_Status

```python
ax=sns.countplot(x="Marital_Status" , hue= "Gender" , data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(x="Marital_Status" , data=df )
for bar in ax.containers:
    ax.bar_label(bars)
```

```python
grouping=df.groupby(["Marital_Status" , "Gender"],as_index=False)["Amount"].sum().sort_values(by = "Amount")

sns.barplot(x="Marital_Status" , y="Amount" , hue= "Gender", data=grouping)
```

#### From the above graph we can observe that most of the buyers are married woman  and they have high purchasing power

## Occupation

```python
ax=sns.countplot(x="Occupation" , data=df )
sns.set(rc={'figure.figsize' :(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(["Occupation"])["Amount"].sum().sort_values( ascending=False)
```

```python
ax=df.groupby(["Occupation"], as_index= False)["Amount"].sum().sort_values(by = "Amount" ,ascending=False)
sns.barplot(x="Occupation" , y="Amount" , data = ax)
```

#### From the above graph we can observe that most of the buryers are from IT , Healthcare , Aviation Sector 

## Product_Category 

```python
ax=sns.countplot(x="Product_Category" , data=df)
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(["Product_Category"])["Amount"].sum().sort_values(ascending=False)
```

```python
ag=df.groupby(["Product_Category"], as_index= False)['Amount'].sum().sort_values(by = 'Amount' , ascending= False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x="Product_Category" ,y ='Amount' , data= ag)
```
